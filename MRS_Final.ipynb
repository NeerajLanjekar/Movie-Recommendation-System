```python
import pandas as pd
from scipy import sparse
pd.set_option('display.notebook_repr_html', True)
```

```python
ratings = pd.read_csv('dataset/ratings.csv')
movies = pd.read_csv('dataset/movies.csv')
links = pd.read_csv('dataset/links.csv')

c = '''
movies = pd.merge(movies,links).drop(['imdbId'],axis=1)
ratings = pd.merge(movies,ratings).drop(['genres','timestamp'],axis=1)
print(ratings.shape)
ratings.head()
print(type(ratings))
'''



# print(links.shape)
# print(links.shape)
# print(links.dropna().shape)
links = links.fillna(0)
# print(links.shape)


links = pd.concat([links.drop(['imdbId','tmdbId'], axis=1), links['tmdbId'].astype(dtype='uint64')], axis=1)
movies_links = pd.merge(movies, links)
movies = pd.merge(movies, links)
ratings = pd.merge(movies,ratings).drop(['genres','timestamp'],axis=1)
# print(ratings.shape)
# ratings.head()
# print(ratings.shape)
# ratings[ratings['tmdbId'] == 0]
ratings = ratings[ratings['tmdbId'] != 0]
print(ratings.shape)
links.head()
movies_links.head()
movies.head()
ratings.head()
```

```python
userRatings = ratings.pivot_table(index=['userId'],columns=['title'],values='rating')
display(userRatings.head())
print("Before: ",userRatings.shape)
userRatings = userRatings.dropna(thresh=10, axis=1).fillna(0,axis=1)
#userRatings.fillna(0, inplace=True)
print("After: ",userRatings.shape)
```

```python
corrMatrix = userRatings.corr(method='pearson')
corrMatrix.head(100)
```

```python
def get_similar(movie_name,rating):
    similar_ratings = corrMatrix[movie_name]*(rating-2.5)
    similar_ratings = similar_ratings.sort_values(ascending=False)
    # print(type(similar_ratings))
    # print(similar_ratings)
    return similar_ratings
```

```python
romantic_lover = [("(500) Days of Summer (2009)",5),("Alice in Wonderland (2010)",3),("Aliens (1986)",1),("2001: A Space Odyssey (1968)",2)]
similar_movies = pd.DataFrame()
for movie,rating in romantic_lover:
    get_similar(movie,rating)
```

```python

def recommendusingcorrelation(rate_tup):
    print(rate_tup)
    similar_movies = pd.DataFrame()
    for movie,rating in rate_tup:
        similar_movies = pd.concat([similar_movies, get_similar(movie,rating)],ignore_index = True, axis=1)
    # similar_movies.head(10)
    m_list = similar_movies.sum(axis=1).sort_values(ascending=False).head(20).index
    print(m_list)
    m_ids = movies_links[movies_links['title'].isin(m_list)]['tmdbId'].values
    return m_ids
```
